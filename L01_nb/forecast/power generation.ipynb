```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten, GlobalAveragePooling1D, LSTM, SimpleRNN
from keras.optimizers import SGD

# Train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)

# Train the XGBoost model (using the previously defined xgb)
xgb=XGBRegressor(n_estimators=100,max_depth=7,learning_rate=0.1)
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)

# Reshape the data for the LSTM model
x_train_lstm = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_test_lstm = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
x_val_lstm = np.reshape(x_val, (x_val.shape[0], x_val.shape[1], 1))

# Redefine the LSTM model
model = Sequential()
# Time series model
model.add(LSTM(64, return_sequences=True, input_shape=(x_train_lstm.shape[1], x_train_lstm.shape[2])))
model.add(Dropout(0.5))

# Feature Extraction
model.add(Conv1D(16, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

model.add(Flatten())
# ANN
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam',
              loss='mse',
              metrics=['r2_score'])

# Train the LSTM model
history = model.fit(x_train_lstm, y_train,
                  validation_data=(x_val_lstm, y_val),
                  epochs=200,
                  batch_size=128,
                  verbose=0) # Set verbose to 0 to avoid printing epoch details


y_pred_lstm = model.predict(x_test_lstm)


# Create the hybrid prediction by averaging the predictions
y_pred_hybrid = (y_pred_rf + y_pred_xgb + y_pred_lstm.flatten()) / 3

# Evaluate the hybrid model
print('Hybrid Model Performance:')
print('RMSE: {:.2f} %'.format(root_mean_squared_error(y_test, y_pred_hybrid)))
print('R2: {:.2f} %'.format(r2_score(y_test, y_pred_hybrid) * 100))
print('MAE: {:.2f} %'.format(mean_absolute_error(y_test, y_pred_hybrid) * 100))
```
